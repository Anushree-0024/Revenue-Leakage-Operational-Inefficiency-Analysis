```python
df['cancelled_by_customer'] = df['Canceled_Rides_by_Customer'].notna()
df['cancelled_by_driver'] = df['Canceled_Rides_by_Driver'].notna()
#currently we are not interested in reason stated by the drivers or customers so lets convert them into boolean values only
```

```python
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
df['Time']=pd.to_datetime(df['Time'],format='%H:%M:%S',errors='coerce')
```

```python
#create a new booking_datetime column
df['booking_datetime']= (
    df['Date'].dt.date.astype(str) + ' ' + df['Time'].dt.time.astype(str)
)
df['booking_datetime'] = pd.to_datetime(df['booking_datetime'], errors='coerce')
```

```python
df[['booking_datetime']]
```

```python

df = df.drop(columns=['Date', 'Time'])
```

```python
df.head()
```

```python
new_column_order = [  # Time
    'booking_datetime',
# Identifiers
    'Booking_ID',
    'Customer_ID',
# Ride details
    'Booking_Status',
    'Vehicle_Type',
    'Pickup_Location',
    'Drop_Location',
# Operational metrics
    'V_TAT',
    'C_TAT',
    'Ride_Distance',
    'Booking_Value',
# Experience metrics
    'Payment_Method',
    'Driver_Ratings',
    'Customer_Rating',
# Cancellation / funnel flags
    'cancelled_by_customer',
    'cancelled_by_driver',
    'is_incomplete',
# Optional diagnostics
    'Incomplete_Rides_Reason'
]
```

```python
df=df[new_column_order]
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.info()
```

```python
#let's create one final conclusion column which we directly tell us whether the ride was cancelled by customer or cancelled by driver or completed or incomplete
df['ride_outcome']=np.select(
    [
        df['cancelled_by_customer'],df['cancelled_by_driver'],df['is_incomplete']
    ],['cancelled_by_customer','cancelled_by_driver','incomplete_ride'],
    default='completed_ride'
    
)
```

```python
df['ride_outcome'].value_counts()
```

```python
#Since the data is for one month only lets break the booking_datetime column into 4 columns to analyse the trend
df['booking_date']=df['booking_datetime'].dt.date
df['booking_hour']=df['booking_datetime'].dt.hour
df['booking_day']=df['booking_datetime'].dt.day_name()
df['is_weekend']=df['booking_day'].isin(['Saturday','Sunday'])
```

```python
df.info()
```

```python
df=df.drop(columns=['cancelled_by_customer','cancelled_by_driver','is_incomplete'])
```

```python
df = df[[
    'booking_datetime',
    'booking_date',
    'booking_hour',
    'booking_day',
    'is_weekend',
    'Booking_ID',
    'Customer_ID',
    'ride_outcome',
    'Vehicle_Type',
    'Pickup_Location',
    'Drop_Location',
    'V_TAT',
    'C_TAT',
    'Ride_Distance',
    'Booking_Value',
    'Payment_Method',
    'Driver_Ratings',
    'Customer_Rating',
    'Incomplete_Rides_Reason'
]]
```

```python
df.to_csv("rides_cleaned_july.csv",index=False)
```
